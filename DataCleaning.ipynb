```python
# we have different lat and lng for same stations, which does not make a good work while grouping data by stations using lat and lng in visuals
# so, we assign avg of all the lat and lng for a station respectively
# separately for start and end stations
Start_Station_Lat=df_oct_nov_nona_clear.groupby('start_station_name')['start_lat'].mean()
Start_Station_Lng=df_oct_nov_nona_clear.groupby('start_station_name')['start_lng'].mean()
Start_Station_groupped=pd.DataFrame(Start_Station_Lat)
Start_Station_groupped=pd.merge(Start_Station_groupped,Start_Station_Lng, how='left', on='start_station_name')
Start_Station_groupped=Start_Station_groupped.dropna()
Start_Station_groupped.head()
```

```python
end_station_Lat=df_oct_nov_nona_clear.groupby('end_station_name')['end_lat'].mean()
end_station_Lng=df_oct_nov_nona_clear.groupby('end_station_name')['end_lng'].mean()
end_station_groupped=pd.DataFrame(end_station_Lat)
end_station_groupped=pd.merge(end_station_groupped,end_station_Lng, how='left', on='end_station_name')
end_station_groupped=end_station_groupped.dropna()
end_station_groupped.head()
```

```python
df_oct_nov_nona_clear=df_oct_nov_nona_clear.drop(['start_lat','start_lng','end_lat', 'end_lng'], axis=1)
df_oct_nov_nona_clear.head()
```

```python
# find those rows that have info about rides in the first part of October 
# in order to find zip codes for those stations
df_zip= df_oct_nov_nona_clear.copy()
df_zip=df_zip[df_zip['ended_at']<pd.Timestamp('2023-10-15')]
df_zip['zip code']=''
df_zip.head()
```

```python
df_zip.shape
```

```python
# https://gis.stackexchange.com/questions/352961/converting-lat-lon-to-postal-code-using-python
import geopy
from geopy.extra.rate_limiter import RateLimiter

# function looks for a zip code of a station based on its lat and lng
def get_zipcode(df, geolocator, lat_field, lon_field):
    # apply wait time of 1 sec before each next record
    geocode = RateLimiter(geolocator.reverse, min_delay_seconds=1)
    
    for index, row in df.iterrows():
        try:
            # get the location details
            location = geocode((row[lat_field], row[lon_field]))
            # get zip code for a location
            df.loc[index, 'zip code'] = location.raw['address'].get('postcode', None)
        except Exception as e:
            # log errors if any
            print(e)
            df.loc[index, 'zip code'] = None
    return df

#  create an instance of the Nominatim geocoder
geolocator = geopy.Nominatim(user_agent='my-geopy-application')

# Call the function for assigning a zip code
df_zip = get_zipcode(df_zip, geolocator, 'start_lat', 'start_lng')   
```

```python
# Get rid of rows with empty zip codes
df_zip = df_zip.dropna()   
df_zip.head()
```

```python
# keep data with zip codes in separate file fo simplisity
df_zip.to_csv('union_oct_nov_2023_zip.csv')
```
